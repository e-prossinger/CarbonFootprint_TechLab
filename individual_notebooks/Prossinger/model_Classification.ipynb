# 1. first (failed) attempt
people might ask themself "is my Carbon Emission good or bad?"

The concept of "Earths" refers to how many planet Earths would be needed to sustain humanity if everyone lived like a specific person. It measures the ecological footprint of an individual's lifestyle, including carbon emissions, resource consumption, and waste production.

ChatGPT says that to live within the means of "one Earth," an individual's annual CO₂ emissions should be 285.25 kg monthly.

```python
#create new column
value_good_CE = 285.25
df['is_CarbonEmission_OK'] = df['CarbonEmission'] < value_good_CE


df[['CarbonEmission','is_CarbonEmission_OK']]
```

```python

# Create Pie-Chart
counts = df['is_CarbonEmission_OK'].value_counts()
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)  # Pie chart
counts.plot.pie(autopct='%1.1f%%',  colors=['lightblue', 'lightgreen'], startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('How many people')
plt.ylabel("")


# Create Histogram
plt.subplot(1, 2, 2)  # Histogram
plt.hist(df['CarbonEmission'], bins=30, color='lightgray', edgecolor='black')
plt.axvline(value_good_CE, color='red', linestyle='dashed', linewidth=2)  # Line for the value_good_CE
plt.title('Histogram of Carbon Emission')
plt.xlabel('Carbon Emission')
plt.ylabel('How many people')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
